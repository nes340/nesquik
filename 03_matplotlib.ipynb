```python
import matplotlib.pyplot as plt
```

```python
# 라인 그래프
# plot(): 인자값을 하나만 넣는 경우 y측에 데이터가 입력되어 그래프 출력
    # 인자값이 2개인 경우 첫 번째 인자값은 x축, 두 번째 인자값은 y축

plt.plot([1,2,3,4])     # x축은 0,1,2,3
plt.show()
```

```python
plt.plot([1,2,3,4],[4,3,7,30])
```

```python
# 바형 그래프
# bar(): x축 데이터와 y축 데이터가 필수 항목
plt.bar([1,2,3],[80,90,60])
plt.show
```

```python
# 산점도 그래프 
import numpy as np 
x = np.random.rand(50)
y = np.random.rand(50)
print(x)
```

```python
# scatter: x축과 y축 데이터가 필수 항목
plt.scatter(x,y)
plt.show()
```
